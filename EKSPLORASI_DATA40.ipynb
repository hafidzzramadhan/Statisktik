```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca file CSV
file_path = "/content/sample_data/houseprices (1).csv"
data = pd.read_csv(file_path)

# Ambil kolom bathrooms dan house prices
bathrooms = data['Bathrooms']
house_prices = data['Price']

# Membuat scatter plot
plt.scatter(bathrooms, house_prices, color='red')

# Memberikan judul dan label sumbu
plt.title('Scatter Plot House Price vs Bathrooms')
plt.xlabel('Bathrooms')
plt.ylabel('House Price')

# Menampilkan plot
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca file CSV
file_path = "/content/sample_data/houseprices (1).csv"
data = pd.read_csv(file_path)

# Group data berdasarkan jumlah kamar mandi (bathrooms)
grouped_data = data.groupby('Bathrooms')['Price'].apply(list)

# Membuat grouped boxplot
plt.figure(figsize=(10, 6))
plt.boxplot(grouped_data.values, labels=grouped_data.index)

# Memberikan judul dan label sumbu
plt.title('Grouped Boxplot House Price by Bathrooms')
plt.xlabel('Bathrooms')
plt.ylabel('House Price')

# Menampilkan plot
plt.grid(True)
plt.show()
```
