```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Memanggil dataset
data_nama = pd.read_csv('/content/sample_data/titanic.csv')

# Mengambil data pada kolom tertentu
data2 = data_nama[['Age', 'Pclass', 'Survived']]

# Menghilangkan baris dengan nilai NaN di kolom 'Age' untuk mencegah error pada plot
data2 = data2.dropna(subset=['Age'])

# Memvisualisasikan data dengan scatter plot
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(data=data2, x='Age', y='Pclass', hue='Survived', palette='Paired')

# Menambahkan judul dan label sumbu
scatter_plot.set_title('Persebaran Data Titanic Berdasarkan Umur, Kelas, dan Keberhasilan Selamat')
scatter_plot.set_xlabel('Age')
scatter_plot.set_ylabel('Pclass')

# Menampilkan plot
plt.show()
```

```python
import pandas as pd

# Memanggil dataset
data_nama = pd.read_csv('/content/sample_data/titanic.csv')

# Memilih kolom tertentu
data3 = data_nama[['Name', 'Sex', 'Age', 'Pclass', 'Fare,,']]

# Mengelompokkan data berdasarkan 'Pclass' dan menghitung jumlah penumpang unik
penumpang = data3.groupby('Pclass')['Name'].nunique()

# Menampilkan jumlah penumpang per kelas
print('Jumlah Penumpang per Kelas:\n', penumpang)
```

```python
import pandas as pd

# Memanggil dataset
data_nama = pd.read_csv('/content/sample_data/titanic.csv')

# Memilih kolom tertentu
data4 = data_nama[['Name', 'Sex', 'Age', 'Pclass', 'Fare,,']]

# Memfilter data penumpang yang tidak selamat berdasarkan Pclass
notsurvivedpassenger = data4['Pclass'].loc[data_nama['Survived'] == 0]
print('Penumpang yang tidak selamat berdasarkan kelas:\n', notsurvivedpassenger.value_counts())

# Memfilter data penumpang yang selamat berdasarkan Pclass
survivedpassenger = data4['Pclass'].loc[data_nama['Survived'] == 1]
print('\nPenumpang yang selamat berdasarkan kelas:\n', survivedpassenger.value_counts())




```

```python
import pandas as pd

# Memanggil dataset
data_nama = pd.read_csv('/content/sample_data/titanic.csv')

# Mengelompokkan data berdasarkan 'Sex' dan menghitung jumlah penumpang
penumpang_per_sex = data_nama.groupby('Sex')['Name'].size()

# Menampilkan jumlah penumpang per jenis kelamin
print('Jumlah Penumpang berdasarkan Jenis Kelamin:\n', penumpang_per_sex)
```

```python
import pandas as pd

# Memanggil dataset
data_nama = pd.read_csv('/content/sample_data/titanic.csv')

# Memilih kolom tertentu
data4 = data_nama[['Name', 'Sex', 'Age', 'Pclass', 'Fare,,', 'Survived']]

# Memfilter data penumpang yang selamat berdasarkan jenis kelamin
selamat_laki_laki = data4.loc[(data4['Survived'] == 1) & (data4['Sex'] == 'male')]
selamat_perempuan = data4.loc[(data4['Survived'] == 1) & (data4['Sex'] == 'female')]

# Menampilkan jumlah penumpang yang selamat berdasarkan jenis kelamin
print('Jumlah Penumpang Laki-laki yang Selamat:', len(selamat_laki_laki))
print('Jumlah Penumpang Perempuan yang Selamat:', len(selamat_perempuan))

# Jika ingin melihat data secara keseluruhan, Anda bisa menggunakan print(selamat_laki_laki) dan print(selamat_perempuan)
```
